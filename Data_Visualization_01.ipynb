# DATA VISUALIZATION

- Data visualization is a technique to represent the data form of Graphs, Plots and Charts.

- There are majorly 3 libraries used for Data Visualization in python:
    1. Matplotlib (Math Plotting)
    2. Seabon (Stats Plotting)
    3. Plotly (Hover Library or Dashboarding)

# Matplotlib

- It is a Data Visualization library in Pyhton
- Used for Mathematical Plotting
- With the help of Matplotlib we can also do resizing of a image in python.
- We basically use a sub-module or sub-library from matplotlib named as pyplot.

```python
import matplotlib.pyplot as plt
```

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('house pricing delhi- MagicBricks.csv')
```

```python
df
```

```python
# LINEPLOT...

# Line plot needs 2 axis (x and y ) both numeric that will be more preferred.

x = np.array([1,2,3,4,5])
y = np.array([10,20,30,40,50])
```

```python
x,y
```

```python
# Create a line plot with these 2 array

plt.plot(x,y)
```

```python
#  We can use markers to reflect the data points or elements

plt.plot(x,y,marker='o')
```

```python
# Plotting data ona Line Plot

plt.plot(df.Area,df.Price)
```

```python
plt.plot(df.Area,df.Price,marker='o')
```

```python
plt.plot(x,y,marker='o',linestyle='dashed')
```

```python
plt.plot(x,y,marker='o',linestyle='-')
```

```python
plt.plot(x,y,marker='o',linestyle='--')
```

```python
# Data Visualization with Data Filteration...

df[df.Location=='Karol Bagh']
```

```python
plt.plot(df[df.Location=='Karol Bagh']['Area'],df[df.Location=='Karol Bagh']['Price'],marker='o',ms=8)
```

```python
# It will resize the figure demensional like lenght and width

plt.figure(figsize=(10,3))

plt.plot(df[df.Location=='Karol Bagh']['Area'],df[df.Location=='Karol Bagh']['Price'],marker='o',ms=8)
```

```python
# In Matplotlib there is no axis given we've to provide the lables accordingly.

plt.figure(figsize=(10,3))
plt.plot(df.Area,df.Price,marker='o')
plt.xlabel('AREA')
plt.ylabel('PRICE')
plt.title('DELHI HOMES AREA and PRICE')
```

```python
# Scatter Plot

plt.scatter(df.Area,df.Price)
```

```python
# We can create comparision plots with scatter

maruti_speed = np.array([100,120,140,160,180])
maruti_avg = np.array([18,16,15,12,10])

ferrari_speed = np.array([250,270,300,350])
ferrari_avg = np.array([12,7,6,5])
```

```python
plt.figure(figsize=(10,4))
plt.scatter(maruti_speed,maruti_avg)
plt.scatter(ferrari_speed,ferrari_avg)
plt.xlabel('AREA')
plt.ylabel('PRICE')
plt.title('DELHI HOMES AREA and PRICE')
```

```python
plt.figure(figsize=(10,4))
plt.scatter(maruti_speed,maruti_avg)
plt.scatter(ferrari_speed,ferrari_avg)
plt.xlabel('AREA')
plt.ylabel('PRICE')
plt.title('DELHI HOMES AREA and PRICE')
plt.grid()
```

```python
# Comparision karol bagh area and price vs pachim vihar area and price.

kb_area = df[df.Location=='Karol Bagh'].Area
kb_price = df[df.Location=='Karol Bagh'].Price

pv_area = df[df.Location=='Paschim Vihar'].Area
pv_price = df[df.Location=='Paschim Vihar'].Price
```

```python
plt.figure(figsize=(10,4))
plt.scatter(kb_area,kb_price)
plt.scatter(pv_area,pv_price)
```

```python
# Barplot

# Barplot is generate on a categorical data.

df.District.value_counts()
```

```python
plt.figure(figsize=(16,4))
col = ['g','orange','y','r','b','k','hotpink','purple']
plt.bar(df.District.value_counts().index,df.District.value_counts(),color=col)
plt.xlabel('Delhi District')
plt.ylabel('Total Availble Homes')
```

```python
plt.figure(figsize=(16,4))
col = ['g','orange','y','r','b','k','hotpink','purple']
plt.barh(df.District.value_counts().index,df.District.value_counts(),color=col)
plt.xlabel('Delhi District')
plt.ylabel('Total Availble Homes')
```

```python
# Histogram in Matplotlib

# Histogram will show us the data distribution
# It will only work with singe column at a time. (Continious Numeric)

plt.hist(df.Price)
```
