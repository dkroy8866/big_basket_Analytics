```python
# matplotlib Pie Plot.

# Pie Chart can be created with the categorical data...

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('house pricing delhi- MagicBricks.csv')
```

```python
# Pie chart can be create with categorical columns(like furnishing)

plt.pie(df.Furnishing.value_counts(),labels=df.Furnishing.value_counts().index,autopct='%1.1f%%',colors=['g','orange','r'],explode=[0.1,0,0],shadow=True)
plt.show()
```

```python
# Boxplot --->>> It usually works with numeric data (continous numeric data)

plt.boxplot(df.Price)
plt.show()
```

# SEABORN

Seaborn stands for statsical visulization. It provides better interface and visualization then matplotlib, but for resizing it uses matplotlib.

```python
import seaborn as sns
```

```python
# Line plot with Seaborn...

plt.figure(figsize=(10,3))
sns.lineplot(x=df.Area,y=df.Price)
plt.title('Price and Area Delhi Homes')
```

```python
# If we want to see the data distribution w.r.t to the categorical columns.

plt.figure(figsize=(10,3))
sns.lineplot(x=df.Area,y=df.Price,hue=df.District)
plt.title('Price and Area Delhi Homes')
```

```python
# Scatter plot with Seaborn...

sns.scatterplot(x=df.Area,y=df.Price,hue=df.Furnishing)
```

```python
# If we want to see the big data points...

sns.scatterplot(x=df.Area,y=df.Price,hue=df.Furnishing,size=df.Price)
```

```python
sns.scatterplot(x=df.Area,y=df.Price,hue=df.Furnishing,size=df.Price,legend=False)
```

```python
sns.scatterplot(x=df.Area,y=df.Price,hue=df.Furnishing,size=df.Price,legend=False,palette='Set1')
```

```python
# Bar Plot with Seaborn...

sns.barplot(x=df.Furnishing.value_counts().index,y=df.Furnishing.value_counts())
```

```python
sns.barplot(x=df.Furnishing.value_counts(),y=df.Furnishing.value_counts().index)
```

```python
# Count Plot

sns.countplot(x=df.Furnishing)
```

```python
# Histograms

# It will show the data distribution mathematically...
# Histogram will only works with continious numerics

sns.histplot(df.Price)
```

```python
# Data Distribution...
# We cannot resize this at all

sns.displot(df.Price)
```

```python
sns.distplot(df.Price)
```

```python
# KDE Plot (Kernel Distribution Estimator)

sns.kdeplot(x=df.Price,fill=True,color='r',hue=df.Furnishing)
```

```python
sns.kdeplot(x=df.Price,fill=True,color='r',hue=df.Type)
```

```python
# Boxplot

sns.boxplot(df.Area,orient='h')
```

```python
# Stripplots and swarmplots

# Striplot will works as a barplot but with datapoints in it...
# X axis has to be categorical and y has to be continous numeric.

sns.stripplot(x=df.Furnishing,y=df.Price)
```

```python
# Swarmplot was also the as stripplot but it will individually plot the data points (Doen't clustm them)

sns.swarmplot(x=df.Furnishing,y=df.Price)
```

```python
# FacetGrid...(Homework)

a=sns.FacetGrid(df,col='Furnishing',row='BHK')
a.map(sns.scatterplot,'Price','Area')
```

```python
# plot for correlation...

df.corr(numeric_only=True)
```

```python
# If we want to visualize the correlation we need to create a HeatMap...

sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='Greens')
```

```python
# We use Pairplot for scatter all thecolumns accordingly with each other.

sns.pairplot(df,hue='Furnishing')
```

```python
# Seaborn's Catplot

sns.catplot(x=df.Furnishing,y=df.Area)
```
